```python
from PIL import Image
from lavis.models import load_model_and_preprocess
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python

def zero_shot_classification(img,text):
    """
    returns the label with the highest match
    """
    model, vis_processors, txt_processors = load_model_and_preprocess(name="clip_feature_extractor", model_type="ViT-B-16", is_eval=True, device=device)
    image = vis_processors["eval"](img).unsqueeze(0).to(device)
    text_input = txt_processors["eval"](text)
    sample = {"image": image, "text_input": [text_input]}
    features = model.extract_features(sample)
    #print(features.image_embeds_proj.shape)

    similarity = features.image_embeds_proj[0,:] @ features.text_embeds_proj[0,:]
    return similarity.cpu().detach().numpy()
```

```python
#upload your image and set the path
test_image = Image.open("/content/download.jpeg").convert("RGB")
label_list=["a blue colored tshirt","a tshirt","a blue hoodie","a cap","a red colored tshirt"]
for label in label_list:
    print("label: ",label," score: ",zero_shot_classification(test_image,label))

#ofcourse it could be modified/improved, but we just want to test it
```
